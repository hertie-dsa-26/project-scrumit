```python
from pathlib import Path
import json
from collections import Counter

DATA_PATH = Path("../../data/raw/entities.ftm.json")

property_counter = Counter()
schema_counter = Counter()
total_rows = 0

with open(DATA_PATH, "r", encoding="utf-8") as f:
    for line in f:
        record = json.loads(line)
        total_rows += 1
        
        schema = record.get("schema")
        if schema is not None:
            schema_counter[schema] += 1
        
        properties = record.get("properties", {})
        property_counter.update(properties.keys())

print("Total rows:", total_rows)
print("Total unique property keys:", len(property_counter))
print("\nTop 30 most common property keys:")
for key, count in property_counter.most_common(30):
    print(f"{key}: {count}")
```

```python
#inspct the propertis as table

property_summary = (
    pd.DataFrame(property_counter.items(), columns=["property", "count"])
    .sort_values("count", ascending=False)
    .reset_index(drop=True)
)

property_summary["share_pct"] = property_summary["count"] / total_rows * 100
property_summary.head(30)
```

```python
#inspect schemas

schema_summary = (
    pd.DataFrame(schema_counter.items(), columns=["schema", "count"])
    .sort_values("count", ascending=False)
    .reset_index(drop=True)
)

schema_summary["share_pct"] = schema_summary["count"] / total_rows * 100
schema_summary.head(20)
```

```python
#keeping those properties that comes 1% of rows
common_properties = property_summary[property_summary["share_pct"] >= 1]["property"].tolist()
common_properties[:50]
```

```python
#defining a flattening function
def flatten_record(record):
    properties = record.get("properties", {})
    
    flat = {
        "id": record.get("id"),
        "caption": record.get("caption"),
        "schema": record.get("schema"),
        "first_seen": record.get("first_seen"),
        "last_seen": record.get("last_seen"),
        "last_change": record.get("last_change"),
        "target": record.get("target"),
        "datasets": ", ".join(record.get("datasets", [])) if record.get("datasets") else None,
        "referents_count": len(record.get("referents", [])) if record.get("referents") else 0,
    }
    
    # Flatten selected properties
    for key, value in properties.items():
        if isinstance(value, list):
            flat[key] = ", ".join(str(v) for v in value) if value else None
        else:
            flat[key] = value
    
    return flat
```

```python
#flattening a small sample first

rows = []

with open(DATA_PATH, "r", encoding="utf-8") as f:
    for i, line in enumerate(f):
        record = json.loads(line)
        rows.append(flatten_record(record))

        if i == 999:   # first 1000 rows
            break

df_sample = pd.DataFrame(rows)

print(df_sample.shape)
df_sample.head()

#inspect variables
#df_sample.columns
```
